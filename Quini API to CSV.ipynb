```python
# Dependencies
import requests
import json
import pandas as pd
from config import api_key
from pprint import pprint
```

```python
# URL for GET requests to retrieve vehicle data
# Note that the ?t= is a query param for the t-itle of the
# movie we want to search for.
base_url = "https://quiniwine.com/api/pub/wineCategory/"
api_key = "&apikey=fYeSO-ChogAFnqAvDHv1" #not needed

limit = 250
categorycode = ['t', 'm', 'p']
```

```python
# Retrieving data and converting it into JSON

# pprint(data)

name =[]
winery = []
area = []
province = []
country = []
varietal = []
vintage = []
style = []
wine_type = []
wine_id = []
rating = []

#when items list is empty, stop making api requests

for y in categorycode:
    
    skip = 0
    
    response = {'items': ['a']}
    
    print(y)

    while len(response['items'])  != 0: #skip < 1000:
        
        print(skip)

        url = base_url + y + '/' + str(skip) + '/' + str(limit)

        response = requests.get(url).json()

        data = response['items']

        for x in data:
            try:
                name.append(x['Name'])
            except:
                name.append('N/A')
            try:
                winery.append(x['Winery'])
            except:
                winery.append('N/A')
            try:
                area.append(x['Area'])
            except:
                area.append('N/A')
            try:
                province.append(x['Province'])
            except:
                province.append('N/A')
            try:
                country.append(x['Country'])
            except:
                country.append('N/A')
            try:
                varietal.append(x['Varietal'])
            except:
                varietal.append('N/A')
            try:
                vintage.append(x['vintage'])
            except:
                vintage.append('N/A')
            try:
                style.append(x['Style'])
            except:
                style.append('N/A')
            try:
                wine_type.append(x['Type'])
            except:
                wine_type.append('N/A')
            try:
                wine_id.append(x['id'])
            except:
                wine_id.append('N/A')
            try:
                rating.append(x['rating'])
            except:
                rating.append('N/A')

        skip = response['nextskip']
```

```python
df = pd.DataFrame({'Name': name, 'Rating': rating, 'Winery': winery, 'Area': area, 'Province': province, 'Country': country,
                   'Varietal': varietal, 'Vintage': vintage, 'Style': style, 'Wine Type': wine_type, 'ID': wine_id})

unique = df.drop_duplicates(subset=['ID'])

unique.to_csv("Wine_data.csv", index=False, header=True)

# df.head()

# print(len(df))

# print(len(df['ID'].unique()))

# print(len(df) - len(df['ID'].unique()))

print(len(unique['ID']))
unique.head()
```
